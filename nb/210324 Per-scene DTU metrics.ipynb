```python
import os
import glob
import pandas as pd
from IPython.display import display
```

```python
eval_out_dir = "/home/ajay/sharedajay/clip/nerf/pixel-nerf/eval_out/"

metric_patterns = {
    "pixelNeRF": os.path.join(eval_out_dir, "dtu_1v/finish.txt"),
    "pixelNeRF, ft": os.path.join(eval_out_dir, "pn*_dtu_ftall1v_lr1e-5_scan*/finish.txt"),
    "pixelDietNeRF": os.path.join(eval_out_dir, "pn*_dtu_ftall1v_lr1e-5_ctr_scale.4_scan*/finish.txt"),
}

metric_paths = {k: sorted(glob.glob(pattern)) for k, pattern in metric_patterns.items()}
```

```python
method_metrics = []

for method, paths in metric_paths.items():
    for path in sorted(paths):
        if "_scan21_src" in path:
            continue
        with open(path, "r") as f:
            for scene_metrics in f:
                scene_metrics = scene_metrics.strip().split()
                if scene_metrics:
                    scan, psnr, ssim, _ = scene_metrics
                    method_metrics.append({
                        "Method": method,
                        "Scan": scan,
                        "PSNR": float(psnr),
                        "SSIM": float(ssim)
                    })

method_metrics = pd.DataFrame(method_metrics)

display(method_metrics)

for key, df in sorted(method_metrics.groupby(["Scan"])):
    display(df)
```
